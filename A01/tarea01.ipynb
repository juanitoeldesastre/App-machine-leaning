```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier

# Datos: (0 = NO, 1 = SÍ)
X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])

# Entrenar el modelo
mlp = MLPClassifier(hidden_layer_sizes=(5,), max_iter=3000, random_state=42)
mlp.fit(X, y)

# Rango para predecir y graficar la curva
x_range = np.linspace(0, 7, 300).reshape(-1, 1)
y_pred_prob = mlp.predict_proba(x_range)[:, 1]  # Probabilidad de clase 1

# Gráfico
plt.figure(figsize=(8, 5))
# Datos originales
plt.scatter(X, y, color='blue', label='Datos originales')
# Curva de predicción
plt.plot(x_range, y_pred_prob, color='red', label='Probabilidad predicha (Clase 1)')
# Línea de umbral
plt.axhline(0.5, color='gray', linestyle='--', label='Umbral 0.5')

plt.title('Clasificación binaria con MLPClassifier')
plt.xlabel('Valor de entrada (X)')
plt.ylabel('Probabilidad de clase 1')
plt.legend()
plt.grid(True)
plt.show()
```
